Second call: Now a is 2. Calling b() again does a = c(a), or a = c(2). The function c(2) returns 2 + 2, so a is now 4.
Third call: Now a is 4. Calling b() again means a = c(a), or a = c(4). The function c(4) returns 4 + 2, so a becomes 6.

As shown above, if the file cannot be found by the interpreter or if it cannot be read
as a text file, an exception will be raised. Modify function fileLength() so that a
friendly message is printed instead:
>>> fileLength('midterm.py')
358

```python
def calculateFileLength(file_name):
    try:
        with open(file_name, 'r') as file_handle:
            file_content = file_handle.read()
        return len(file_content)
    except FileNotFoundError:
        print(f"File {file_name} not found.")

# Testing the modified function
print(calculateFileLength('midterm.py')) 
print(calculateFileLength('idterm.py')) 
```

#question 3

Now write a class named Kangaroo as a subclass of Marsupial that inherits all the
attributes of Marsupial and also:
a. extends the Marsupial __init__ constructor to take, as input, the
coordinates x and y of the Kangaroo object,
b. supports method jump that takes number values dx and dy as input and
moves the kangaroo by dx units along the x-axis and by dy units along the y-
axis, and
c. overloads the __str__ operator so it behaves as shown below.
>>> k = Kangaroo(0,0)
>>> print(k)
I am a Kangaroo located at coordinates (0,0)
>>> k.put_in_pouch('doll')
>>> k.put_in_pouch('firetruck')
>>> k.put_in_pouch('kitten')
>>> k.pouch_contents()
['doll', 'firetruck', 'kitten']
>>> k.jump(1,0)
>>> k.jump(1,0)
>>> k.jump(1,0)
>>> print(k)
I am a Kangaroo located at coordinates (3,0)

```python
class Marsupial:
    def __init__(self):
        # Initialize an empty list to represent the pouch contents
        self.pouch = []

    def put_in_pouch(self, item):
        # Add the specified item to the pouch list
        self.pouch.append(item)

    def pouch_contents(self):
        # Return the current contents of the pouch
        return self.pouch


class Kangaroo(Marsupial):
    def __init__(self, x, y):
        # Call the parent class (Marsupial) constructor to initialize the pouch
        super().__init__()
        # Set the initial x and y coordinates of the Kangaroo
        self.x = x
        self.y = y

    def jump(self, dx, dy):
        # Update the x coordinate by dx and the y coordinate by dy
        self.x += dx
        self.y += dy

    def __str__(self):
        # Return a string representation of the Kangaroo's current location
        return f"I am a Kangaroo located at coordinates ({self.x},{self.y})"


# Example Usage
k = Kangaroo(0, 0)           # Create a Kangaroo at coordinates (0,0)
print(k)                     # Print the current location of the Kangaroo
# Output: I am a Kangaroo located at coordinates (0,0)

k.put_in_pouch('doll')       # Add 'doll' to the pouch
k.put_in_pouch('firetruck')  # Add 'firetruck' to the pouch
k.put_in_pouch('kitten')     # Add 'kitten' to the pouch
print(k.pouch_contents())    # Check the contents of the pouch
# Output: ['doll', 'firetruck', 'kitten']

k.jump(1, 0)                 # Move the Kangaroo by (1, 0)
k.jump(1, 0)                 # Move the Kangaroo by another (1, 0)
k.jump(1, 0)                 # Move the Kangaroo by another (1, 0)
print(k)                     # Print the current location after jumps
# Output: I am a Kangaroo located at coordinates (3,0)
```

#Question 4 

Write function collatz() that takes a positive integer x as input and prints the
Collatz sequence starting at x. A Collatz sequence is obtained by repeatedly applying
this rule to the previous number x in the sequence:
x = { 𝑥/2 𝑖𝑓 𝑥 𝑖𝑠 𝑒𝑣𝑒𝑛
3𝑥 + 1 𝑖𝑓 𝑥 𝑖𝑠 𝑜𝑑𝑑
Your function should stop when the sequence gets to number 1. Your
implementation must be recursive, without any 
>>> collatz(1)
1
>>> collatz(10)
10
5
16
8
4
2
1loops.

```python
def collatz(x):
    # Print the current value of x
    print(x)
    # Base case: if x is 1, stop recursion
    if x == 1:
        return
    # Recursive case: if x is even, call collatz with x / 2
    elif x % 2 == 0:
        collatz(x // 2)
    # If x is odd, call collatz with 3 * x + 1
    else:
        collatz(3 * x + 1)


# Example Usage
print("Collatz sequence for 1:")
collatz(1)   # Expected output: 1

print("\nCollatz sequence for 10:")
collatz(10)  # Expected output: 10, 5, 16, 8, 4, 2, 1
```

#Question 5

Write a recursive method binary() that takes a non-negative
integer n and prints the binary representation of integer n.
>>> binary(0)
0
>>> binary(1)
1
>>> binary(3)
11
>>> binary(9)
1001

```python
def binary(n):
    # Base case: if n is 0 or 1, just print n
    if n < 2:
        print(n, end='')  # Use end='' to avoid newline between digits
    else:
        # Recursive case: call binary with n // 2, then print n % 2
        binary(n // 2)
        print(n % 2, end='')  # Print the remainder to get the binary digit


# Example Usage
print("Binary of 0:")
binary(0)   # Expected output: 0
print("\nBinary of 1:")
binary(1)   # Expected output: 1
print("\nBinary of 3:")
binary(3)   # Expected output: 11
print("\nBinary of 9:")
binary(9)   # Expected output: 1001
```

#Question 6

Implement a class named HeadingParser that can be used to parse an HTML document, and retrieve and print all the headings in the document. You should implement your class as a subclass of HTMLParser, defined in Standard Library module html.parser. When fed a string containing HTML code, your class should print the headings, one per line and in the order in which they appear in the document. Each heading should be indented as follows: an h1 heading should have indentation 0, and h2 heading should have indentation 1, etc. Test your implementation using w3c.html.
>>> infile = open('w3c.html')
>>> content = infile.read()
>>> infile.close()
>>> hp = HeadingParser()
>>> hp.feed(content)
W3C Mission
Principles

```python
from html.parser import HTMLParser

class HeadingParser(HTMLParser):
    def __init__(self):
        super().__init__()
        self.current_depth = 0
    
    def handle_starttag(self, tag, attrs):
        if tag in ('h1', 'h2', 'h3', 'h4', 'h5', 'h6'):
            self.current_depth = int(tag[1])
    
    def handle_endtag(self, tag):
        if tag in ('h1', 'h2', 'h3', 'h4', 'h5', 'h6'):
            self.current_depth = 0
    
    def handle_data(self, data):
        if self.current_depth > 0:
            print(' ' * (self.current_depth - 1) + data.strip())

if __name__ == "__main__":
    infile = open('w3c.html', 'r')
    content = infile.read()
    infile.close()

    hp = HeadingParser()
    hp.feed(content)
```

#Question 7

Implement recursive function webdir() that takes as input: a URL (as a string) and 
non-negative integers depth and indent. Your function should visit every we 
page reachable from the starting URL web page in depth clicks or less, and pri t
each web page's URL. As shown below, indentation, specified by indent, sho ld
be used to indicate the depth of a URL.
>>>
webdir('http://reed.cs.depaul.edu/lperkovic/csc242/test1.html'
, 2, 0)
http://reed.cs.depaul.edu/lperkovic/csc242/test1.html
http://reed.cs.depaul.edu/lperkovic/csc242/test2.html
http://reed.cs.depaul.edu/lperkovic/csc242/test4.html
http://reed.cs.depaul.edu/lperkovic/csc242/test3.html
http://reed.cs.depaul.edu/lperkovic/csc242/test4.html

```python
import requests
from bs4 import BeautifulSoup

def webdir(url, depth, indent):
    # Base case: if depth is less than 0, stop recursion
    if depth < 0:
        return

    # Print the current URL with the specified indentation
    print(' ' * indent + url)

    try:
        # Send a request to the URL
        response = requests.get(url)
        # Parse the HTML content
        soup = BeautifulSoup(response.text, 'html.parser')
    except Exception as e:
        print(f"Could not retrieve or parse {url}: {e}")
        return

    # Find all anchor tags with 'href' attributes
    for link in soup.find_all('a', href=True):
        next_url = link['href']
        # Resolve relative URLs
        if not next_url.startswith('http'):
            next_url = requests.compat.urljoin(url, next_url)
        # Recursive call with reduced depth and increased indentation
        webdir(next_url, depth - 1, indent + 2)

# Prompt the user for the URL, depth, and indentation level
start_url = input("Enter the starting URL: ")
try:
    max_depth = int(input("Enter the depth (non-negative integer): "))
    initial_indent = int(input("Enter the initial indentation level (non-negative integer): "))
except ValueError:
    print("Invalid input. Depth and indentation must be non-negative integers.")
else:
    # Start the webdir function with user input
    webdir(start_url, max_depth, initial_indent)
```

#Question 8

Write SQL queries on the below database table that return:
a) All the temperature data.
b) All the cities, but without repetition.
c) All the records for India.
d) All the Fall records.
e) The city, country, and season for which the average rainfall is between 200
and 400 millimeters.
f) The city and country for which the average Fall temperature is ab ve 20
degrees, in increasing temperature order.
g) The total annual rainfall for Cairo.
h) The total rainfall for each season.


```python
import sqlite3

# Connect to the SQLite database (or create it if it doesn’t exist)
weather_db_conn = sqlite3.connect('weather_database.db')
weather_db_cursor = weather_db_conn.cursor()

# Define the WeatherInfo table structure in the database
weather_db_cursor.execute('''
CREATE TABLE IF NOT EXISTS WeatherInfo (
    CityName TEXT,
    CountryName TEXT,
    SeasonName TEXT,
    AvgTemperature REAL,
    TotalRainfall REAL
)
''')

# Weather data to be inserted into the WeatherInfo table
weather_data_entries = [
    ('Mumbai', 'India', 'Winter', 24.8, 5.9),
    ('Mumbai', 'India', 'Spring', 28.4, 16.2),
    ('Mumbai', 'India', 'Summer', 27.9, 1549.4),
    ('Mumbai', 'India', 'Fall', 27.6, 346.0),
    ('London', 'United Kingdom', 'Winter', 4.2, 207.7),
    ('London', 'United Kingdom', 'Spring', 8.3, 169.6),
    ('London', 'United Kingdom', 'Summer', 15.7, 157.0),
    ('London', 'United Kingdom', 'Fall', 10.4, 218.5),
    ('Cairo', 'Egypt', 'Winter', 13.6, 16.5),
    ('Cairo', 'Egypt', 'Spring', 20.7, 6.5),
    ('Cairo', 'Egypt', 'Summer', 27.7, 0.1),
    ('Cairo', 'Egypt', 'Fall', 22.2, 4.5)
]

# Populate WeatherInfo table with weather data
weather_db_cursor.executemany('''
INSERT INTO WeatherInfo (CityName, CountryName, SeasonName, AvgTemperature, TotalRainfall)
VALUES (?, ?, ?, ?, ?)
''', weather_data_entries)

# Save changes to the database
weather_db_conn.commit()

# Utility function to run and display query results
def execute_query(query_text):
    weather_db_cursor.execute(query_text)
    query_results = weather_db_cursor.fetchall()
    for result in query_results:
        print(result)

# a) Retrieve all temperature records
print("a) Temperature data:")
execute_query("SELECT AvgTemperature FROM WeatherInfo;")
print("\n")

# b) List unique cities only
print("b) Unique cities:")
execute_query("SELECT DISTINCT CityName FROM WeatherInfo;")
print("\n")

# c) Retrieve all records for India
print("c) Records for India:")
execute_query("SELECT * FROM WeatherInfo WHERE CountryName = 'India';")
print("\n")

# d) Retrieve all records for Fall season
print("d) Fall season records:")
execute_query("SELECT * FROM WeatherInfo WHERE SeasonName = 'Fall';")
print("\n")

# e) Retrieve city, country, and season for rainfall between 200 and 400 mm
print("e) City, country, and season with average rainfall between 200 and 400 mm:")
execute_query('''
SELECT CityName, CountryName, SeasonName
FROM WeatherInfo
GROUP BY CityName, CountryName, SeasonName
HAVING AVG(TotalRainfall) BETWEEN 200 AND 400;
''')
print("\n")

# f) List city and country with Fall temperature above 20 degrees, ordered by temperature
print("f) Cities and countries with Fall temperatures above 20 degrees (ordered):")
execute_query('''
SELECT CityName, CountryName
FROM WeatherInfo
WHERE SeasonName = 'Fall'
GROUP BY CityName, CountryName
HAVING AVG(AvgTemperature) > 20
ORDER BY AVG(AvgTemperature) ASC;
''')
print("\n")

# g) Calculate total annual rainfall for Cairo
print("g) Total annual rainfall for Cairo:")
execute_query("SELECT SUM(TotalRainfall) AS TotalAnnualRainfall FROM WeatherInfo WHERE CityName = 'Cairo';")
print("\n")

# h) Calculate total rainfall for each season
print("h) Total rainfall by season:")
execute_query('''
SELECT SeasonName, SUM(TotalRainfall) AS TotalRainfall
FROM WeatherInfo
GROUP BY SeasonName;
''')

# Close the database connection to free resources
weather_db_conn.close()
```
